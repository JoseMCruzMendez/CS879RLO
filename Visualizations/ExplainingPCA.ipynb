```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Gist of explanation: PCA re-centers the origin and rotates so that going in a specific axis corresponds to moving in the direction of greatest increase of covariance
```

```python
#Step 1: Initialize covariance and mean
cov = np.array([[1, 0.1],
                [1.2, 1]])

mean = np.array([0, 0])

# Step 2: Create a grid of (x, y) points
x = np.linspace(-10, 10, 200)
y = np.linspace(-10, 10, 200)
X, Y = np.meshgrid(x, y)

pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X
pos[:, :, 1] = Y
sqrt2 = np.sqrt(2)
#Rotate the distribution for demonstration purposes
rot = np.array([
    [sqrt2/2, sqrt2/2],
    [sqrt2/2, -sqrt2/2]
])
```

```python
#step 3 Create multivariate normal
recentered = pos - mean
inv_cov = np.linalg.inv(cov)
dist = np.einsum("ijk, kk, ijk->ij", recentered, inv_cov, recentered)
dist.shape
```

```python
plt.pcolormesh(X, Y, dist)
```
